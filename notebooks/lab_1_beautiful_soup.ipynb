## Scrapping all albums from Leonard Cohen

```python
url = "https://www.leonardcohenfiles.com/songind.html"
result = requests.get(url)
```

```python
content = result.text
```

```python
soup = BeautifulSoup(content, "lxml")
```

```python
table = soup.find("table")
```

```python
# getting all lines with urls
lines_urls = table.find_all("a", href=True)
```

```python
urls = [u["href"] for u in lines_urls]
```

```python
uniq_urls = np.unique(sorted(urls))
```

```python
proper_urls = ["https://www.leonardcohenfiles.com/" + u for u in  uniq_urls ]
```

```python
url = proper_urls[0]
result = requests.get(url)
content = result.text
```

```python
soup = BeautifulSoup(content, "lxml")
```
